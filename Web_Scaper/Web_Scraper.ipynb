```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
data = pd.read_excel("V2_Query Result Sample.xlsx")
data.head()
```

```python
data.shape
```

```python
working_data = data.iloc[:10, 4:6]
working_data.head()
```

```python
url = "https://www.sec.gov/edgar/searchedgar/companysearch.html"
```

```python
url_list = []
for i in range(len(working_data)):
    driver = webdriver.Chrome("C:/Users/manis/Desktop/chromedriver/chromedriver.exe")
    driver.get(url)

    time.sleep(3)
    searchterm = str(working_data.iloc[i, 0])

    time.sleep(3)
    cik_input = driver.find_element_by_id('cik')
    cik_input.send_keys(searchterm)

    submit = driver.find_element_by_id('cik_find')
    submit.click()

    time.sleep(3)
    file_type = driver.find_element_by_id('type')
    file_type.send_keys('8-K')

    time.sleep(3)
    file_type_submission = driver.find_element_by_xpath('//*[@id="contentDiv"]/div[2]/form/table/tbody/tr/td[6]/input[1]')
    file_type_submission.click()


    time.sleep(3)
    table = driver.find_element_by_xpath("//*[@id='seriesDiv']/table/tbody/tr[2]/td[2]")
    table.click()

    time.sleep(3)
    next_table = driver.find_element_by_class_name('tableFile')

    time.sleep(3)
    # rows = next_table.find_element_by_tag_name('tr')
    rows = next_table.find_elements_by_tag_name('tr')
    # print(len(rows))

    pattern = "Complete submission text file"
    # main_row = rows.find_element_by_class_name('blueRow')
    time.sleep(3)
    flag = False
    for row in rows:
        time.sleep(2)
        cols = row.find_elements_by_tag_name('td')
        for col_index in range(len(cols)):
            if cols[col_index].text == pattern:
                print(cols[col_index + 1].click())
                flag = True
                break
        if flag:
            break

    required_url = driver.current_url
    url_list.append(required_url)
    print(required_url)

    time.sleep(2)
    driver.close()
```
